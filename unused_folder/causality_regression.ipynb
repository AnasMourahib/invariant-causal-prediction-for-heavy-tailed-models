```python

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
```

The causal model $Y = f(X) + E$, with $f(x) = x$

```python
X = np.random.normal(loc = 0 , scale = 1 , size = 1000 )
E = np.random.normal(loc = 0 , scale = 1 , size = 1000 )
Y = X + E

```

We perform a linear regression Y on X

```python
X_reshaped = X.reshape((-1 , 1))
lm = LinearRegression().fit(X_reshaped, Y)
slope = lm.coef_
intercept = lm.intercept_

def estim_fun(x , slope , intercept):
    estim = slope * x + intercept
    return(estim)

estimates_Y = []
for x in X_reshaped :
    estimates_Y.append(estim_fun(x, slope, intercept))
    
    
estimates_Y = np.array(estimates_Y).reshape(-1)    



Residual_Y =  Y - estimates_Y  


plt.scatter(Residual_Y , X)
print(slope)
print(intercept)
print('Covariance between residuals and X:',np.cov(Residual_Y , X)[0][1])

```

We simulate $Y^e = \beta_1 X_1^e + \beta_2 X_2^e + \epsilon^e $, for an environement $e \in E$. Suppose we have two environements $r , \, f \in E$. Assume that both $(X_1^r , X_2^r)$ and $(X_1^f , X_2^f)$ have independent $Pa(1)$ margins, that is $\mathbb{P}(X_i^e > x)  = 1/x, \, x> 1$. Suppose that both $\epsilon^r$ and $\epsilon^f$ follow a $Pa(2)$ distribution, the first with scale parameter $1$ and second with scale parameter $2$. Note that with this setting, the covariates $\boldsymbol{X}$ are heavier than the noise $\epsilon$, on the tail. Note also that across the two environements $r, \, f$, the noise has comparable tails.   

```python
N = 10000
Beta1 = 1 
Beta2 = 2

###Environement r 
X1_r = 1 + np.random.pareto(1, N)  
X2_r = 1 + np.random.pareto(1, N) 
epsilon_r = 1 + np.random.pareto(2, N) 
Y_r = Beta1 * X1_r + Beta2 * X2_r + epsilon_r 
covariates_r  = np.column_stack((X1_r, X2_r))
###Environement f 
X1_f = 1 +  np.random.pareto(1, N)  
X2_f = 1 + np.random.pareto(1, N) 
epsilon_f = (np.random.pareto(2, N) + 1 ) * 2
Y_f = Beta1 * X1_f + Beta2 * X2_f + epsilon_f 
covariates_f  = np.column_stack((X1_f, X2_f))


#covariates.sahpe
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
####For environement r
regr = linear_model.LinearRegression()
regr.fit(covariates_r , Y_r)
Y_r_hat =  (covariates_r * regr.coef_).sum(axis=1)
resid_r = Y_r - Y_r_hat
###For environement f
regr = linear_model.LinearRegression()
regr.fit(covariates_f , Y_f)
Y_f_hat =  (covariates_f * regr.coef_).sum(axis=1)
resid_f = Y_f - Y_f_hat
```

```python
from scipy.stats import genpareto
threshold_r = np.quantile(resid_r, q= 0.95)
exceedances_r = []
for value in resid_r :
    if value > threshold_r:
        exceedances_r.append(value)
exceedances_r = np.array(exceedances_r - threshold_r)
shape_r , loc_r ,  scale_r = genpareto.fit(exceedances_r , floc = 0)

print("For the regular environment: ")
print("The threshold is: ", threshold_r)
print("The shape parameter is: ", shape_r)
print("The scale parameter is:", scale_r)
```

We start by writing the function $ELR$ defined on the third page of the paper
